```python
import requests
import zipfile
import os

# URL del dataset en el UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/static/public/420/autistic+spectrum+disorder+screening+data+for+adolescent.zip"

# Nombre del archivo donde se guardará
file_name = "Autism-Adolescent-Data.arff.zip"

# Descargar el archivo desde la URL
response = requests.get(url)

# Guardar el archivo descargado
with open(file_name, 'wb') as file:
    file.write(response.content)

print(f"Dataset descargado y guardado como {file_name}")

# Obtener el directorio actual
current_directory = os.path.dirname(os.path.abspath(file_name))

# Descomprimir el archivo en la misma carpeta
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall(current_directory)

print(f"Archivo descomprimido en la carpeta {current_directory}")

# Eliminar el archivo .dox si existe
dox_file = os.path.join(current_directory, "/content/Autism-Screening-Adolescent-Data Description.docx")
if os.path.exists(dox_file):
    os.remove(dox_file)
    print(f"Archivo {dox_file} eliminado.")
else:
    print(f"Archivo {dox_file} no encontrado.")
```

# Proceso de Transformacion de los datos:

```python
from sklearn.preprocessing import PowerTransformer, QuantileTransformer, RobustScaler
from sklearn.impute import SimpleImputer

# Asegurarnos de que no haya valores NaN antes de la transformación
def check_numeric_columns(data, numeric_cols):
    """Verifica si todas las columnas numéricas están presentes y son realmente numéricas."""
    missing_cols = set(numeric_cols).difference(data.columns)
    if missing_cols:
        raise ValueError(f"Las siguientes columnas están faltando: {missing_cols}")

    # Verifica que todas las columnas sean numéricas
    for col in numeric_cols:
        if not np.issubdtype(data[col].dtype, np.number):
            raise ValueError(f"La columna {col} no es numérica.")

    # Asegurarse de que no haya NaN
    if data[numeric_cols].isnull().values.any():
        print(f"Valores NaN detectados en las columnas numéricas. Imputando valores...")
        imputer = SimpleImputer(strategy='mean')  # Se puede cambiar a 'median' u otra estrategia
        data[numeric_cols] = imputer.fit_transform(data[numeric_cols])

    return data

# Función para aplicar PowerTransformer
def transform_power(data, numeric_cols):
    """Aplica PowerTransformer a las columnas numéricas de un DataFrame."""
    data = check_numeric_columns(data, numeric_cols)
    pt = PowerTransformer(method='yeo-johnson')
    data_transformed = data.copy()
    data_transformed[numeric_cols] = pt.fit_transform(data[numeric_cols])
    return data_transformed

# Función para aplicar QuantileTransformer
def transform_quantile(data, numeric_cols):
    """Aplica QuantileTransformer a las columnas numéricas de un DataFrame."""
    data = check_numeric_columns(data, numeric_cols)
    qt = QuantileTransformer(output_distribution='normal')
    data_transformed = data.copy()
    data_transformed[numeric_cols] = qt.fit_transform(data[numeric_cols])
    return data_transformed

# Función para aplicar RobustScaler
def transform_robust(data, numeric_cols):
    """Aplica RobustScaler a las columnas numéricas de un DataFrame."""
    data = check_numeric_columns(data, numeric_cols)
    rs = RobustScaler()
    data_transformed = data.copy()
    data_transformed[numeric_cols] = rs.fit_transform(data[numeric_cols])
    return data_transformed

# Asegurarse de que no haya problemas con valores nulos o tipos incorrectos
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Aplicar las transformaciones
datos_entrenamiento_power = transform_power(datos_entrenamiento, numeric_cols)
datos_testing_power = transform_power(datos_testing, numeric_cols)

datos_entrenamiento_quantile = transform_quantile(datos_entrenamiento, numeric_cols)
datos_testing_quantile = transform_quantile(datos_testing, numeric_cols)

datos_entrenamiento_robust = transform_robust(datos_entrenamiento, numeric_cols)
datos_testing_robust = transform_robust(datos_testing, numeric_cols)
```
