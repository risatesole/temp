## 0. Descargar e importar el dataset Dataset

```python
import requests
import zipfile
import os

# URL del dataset en el UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/static/public/420/autistic+spectrum+disorder+screening+data+for+adolescent.zip"

# Nombre del archivo donde se guardará
file_name = "Autism-Adolescent-Data.arff.zip"

# Descargar el archivo desde la URL
response = requests.get(url)

# Guardar el archivo descargado
with open(file_name, 'wb') as file:
    file.write(response.content)

print(f"Dataset descargado y guardado como {file_name}")

# Obtener el directorio actual
current_directory = os.path.dirname(os.path.abspath(file_name))

# Descomprimir el archivo en la misma carpeta
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall(current_directory)

print(f"Archivo descomprimido en la carpeta {current_directory}")

# Eliminar el archivo .dox si existe
dox_file = os.path.join(current_directory, "/content/Autism-Screening-Adolescent-Data Description.docx")
if os.path.exists(dox_file):
    os.remove(dox_file)
    print(f"Archivo {dox_file} eliminado.")
else:
    print(f"Archivo {dox_file} no encontrado.")
```

## 3.Codificacion de variables categorias

```python

'''
Convertir los valores a numeros y unos y ceros:
'''
# columnas autismo a ceros y unos
df['austim'] = df['austim'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

# Map valores de genero
df['austim'] = df['austim'].replace({'no': 0, 'yes': 1})

# df['austim']


# Convertir valores de genero a 0 para male y 1 para femenino
df['gender'] = df['gender'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x) # Convertir cadenas de bytes en cadenas normales

df['gender'] = df['gender'].replace({'m': 0, 'f': 1})

# df['gender']

# convertir columna Class/ASD a ceros y unos
df['Class/ASD'] = df['Class/ASD'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

# Map valores de genero
df['Class/ASD'] = df['Class/ASD'].replace({'NO': 0, 'YES': 1})

df['Class/ASD']

# Convertir age valores en valores enteros
df['age'] = df['age'].astype('Int64')  # Usa 'Int64' para manejar NaN como enteros

# df['age']

# Puntuacion de respuesta:
# Convertir columnas A1_Score a A10_Score de bytes a enteros
for col in ['A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score',
            'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'A10_Score']:
    df[col] = df[col].apply(lambda x: int(x.decode('utf-8')) if isinstance(x, bytes) else int(x))

# Verifica los cambios
df[['A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score']].head()

df
```

# Proceso de Transformacion de los datos:

```python
import pandas as pd
from sklearn.preprocessing import PowerTransformer, QuantileTransformer, RobustScaler

# Supongamos que 'df' es tu DataFrame y 'Class/ASD' es la columna que deseas predecir
# df = pd.read_csv('tu_archivo.csv')  # Cargar tu DataFrame aquí

# Función para aplicar PowerTransformer
def apply_power_transformer(df):
    features = df.drop('Class/ASD', axis=1)  # Características
    pt = PowerTransformer(method='yeo-johnson')
    features_transformed = pt.fit_transform(features)

    # Crear un nuevo DataFrame con las características transformadas
    df_transformed = pd.DataFrame(features_transformed, columns=features.columns)
    # Añadir la columna objetivo si se desea
    df_transformed['Class/ASD'] = df['Class/ASD'].values
    return df_transformed

# Función para aplicar QuantileTransformer
def apply_quantile_transformer(df):
    features = df.drop('Class/ASD', axis=1)  # Características
    qt = QuantileTransformer(output_distribution='normal')
    features_transformed = qt.fit_transform(features)

    # Crear un nuevo DataFrame con las características transformadas
    df_transformed = pd.DataFrame(features_transformed, columns=features.columns)
    # Añadir la columna objetivo si se desea
    df_transformed['Class/ASD'] = df['Class/ASD'].values
    return df_transformed

# Función para aplicar RobustScaler
def apply_robust_scaler(df):
    features = df.drop('Class/ASD', axis=1)  # Características
    rs = RobustScaler()
    features_transformed = rs.fit_transform(features)

    # Crear un nuevo DataFrame con las características transformadas
    df_transformed = pd.DataFrame(features_transformed, columns=features.columns)
    # Añadir la columna objetivo si se desea
    df_transformed['Class/ASD'] = df['Class/ASD'].values
    return df_transformed
```
